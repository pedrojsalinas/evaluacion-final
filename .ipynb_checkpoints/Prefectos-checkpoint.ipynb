```python
import ipywidgets as widgets 
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
```

```python
df = pd.read_csv("final.csv")
```

```python
df.head()
```

```python
provincias = df['provincia'].unique()
w = widgets.Dropdown(
    options=provincias,
    value = u'Napo',
    description='Provincias:',
    disabled=False,
)
display(w)
```

```python
df[df['provincia']== w.value]
```

```python
ax = df.groupby(['provincia']).size().sort_values(ascending=False)[:20].plot.bar(title = u'Mayor cantidad de prefectos según provincia', legend=True)

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
```
